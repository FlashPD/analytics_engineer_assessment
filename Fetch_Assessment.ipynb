# Load data

```python
import pandas as pd
import json
```

## Receipts

```python
# Initialize an empty list to store the JSON objects
receipts_data = []

# Read and parse each line of the JSON file
with open('/content/receipts.json', 'r') as file:
    for line in file:
        receipts_data.append(json.loads(line))
```

```python
# Convert the list of dictionaries to a pandas DataFrame
receipts_df = pd.DataFrame(receipts_data)
```

```python
receipts_df.head()
```

```python
receipts_df.dateScanned.head()
```

```python
date_columns = ['createDate', 'dateScanned', 'finishedDate', 'modifyDate', 'pointsAwardedDate', 'purchaseDate']  # List all column names that contain dates

for column in date_columns:
    # Convert each date column from {'$date': <timestamp>} to datetime format
    # The lambda function extracts the timestamp and divides by 1000 to convert milliseconds to seconds
    receipts_df[column] = pd.to_datetime(receipts_df[column].apply(lambda x: x['$date'] / 1000 if isinstance(x, dict)
    and '$date' in x else x / 1000 if isinstance(x, (int, float)) else x), unit='s')
```

```python
receipts_df.head()
```

```python
# Extract id from the dictionary
receipts_df['_id'] = receipts_df['_id'].apply(lambda x: x['$oid'])
```

```python
receipts_df['_id'].head()
```

```python
# Rename _id column to receipt_id for consistency
receipts_df = receipts_df.rename( columns = {'_id':'receipt_id'})
```

```python
receipts_df.head()
```

```python
# Plot the status of rewards received

from matplotlib import pyplot as plt
import seaborn as sns
receipts_df.groupby('rewardsReceiptStatus').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

## Users

```python
# Initialize an empty list to store the JSON objects
users_data = []

# Read and parse each line of the JSON file
with open('/content/users.json', 'r') as file:
    for line in file:
        users_data.append(json.loads(line))
```

```python
# Convert the list of dictionaries to a pandas DataFrame
users_df = pd.DataFrame(users_data)
```

```python
users_df.head()
```

```python
# Convert date fields from Unix timestamp (milliseconds) to datetime
date_columns = ['createdDate', 'lastLogin']
for column in date_columns:
    users_df[column] = pd.to_datetime(users_df[column].apply(lambda x: x['$date'] / 1000 if isinstance(x, dict) else x), unit='s')
```

```python
users_df.head()
```

```python
# Extract user_id from the dictionary
users_df['_id'] = users_df['_id'].apply(lambda x: x['$oid'])
```

```python
users_df._id.head()
```

```python
# Rename _id to user_id for consistency
users_df = users_df.rename(columns = {'_id':'user_id'})
```

```python
users_df.head()
```

```python
# Plot the number of users by state

import matplotlib.pyplot as plt

# Count the number of users for each state
state_counts = users_df['state'].value_counts()

# Plot a bar chart
state_counts.plot(kind='bar', figsize=(8, 5))
plt.title('Number of Users by State')
plt.xlabel('State')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
users_df.role.unique()
```

```python
# Check number of unique entries in the User table
users_df.user_id.count(),users_df.user_id.nunique()
```

Data quality issue with Users table:

There are more than 57% duplicate entries in the Users table, there seems to be an issue with capturing user data within the app. This could lead to a lot more duplicate data when joining the User table with any other table for analysis, I would suggest to remove the duplicate entries before loading data into the table.

```python
# Drop duplicate user_id's from Users table

users_df = users_df.drop_duplicates(subset = ['user_id'])
```

```python
# Check number of unique user_id's after dropping duplicates
users_df.user_id.count(),users_df.user_id.nunique()
```
